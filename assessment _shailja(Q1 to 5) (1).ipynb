```python
import pandas as pd
from datetime import datetime
independence_day = datetime(year=1947, month=8, day=15)

current_date = datetime.now()

date_range = pd.date_range(start=independence_day, end=current_date, closed='right')

df = pd.DataFrame({
    'day': date_range.day,
    'month': date_range.month,
    'year': date_range.year
})

df.to_csv('date_range.csv', index=False)
print(df)
```

```python
#Question3
def most_k_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num
 
List = [1, 2, 3, 4, 5, 1, 2, 3, 1, 2, 1]
print(most_k_frequent(List))
    
```
